# Repo analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("contributors.csv")

print(df.head())

df.plot()
plt.show()

df['average_watchers'] = df['repo_watchers'] / df['repo_count']

print(df.head())

df = df.sort_values(by='average_watchers', ascending=False)

print(df.head())

df.plot()
plt.show()
```
